# Part 1:  Toronto Segmentation and Clustering of Neighborhoods

This notebook will be used to first build the code to scrape the provided Wikipedia page noted below in order to obtain the data that is in the table of postal codes and to transform the data into a pandas dataframe.

First step is to import required libraries and packages

```python
# import all necessary libraries and packages to be used:  additional libraries / notebooks added in Part 2 & Part 3 when needed.  DJM 6/8/2020
import pandas as pd
import requests
from bs4 import BeautifulSoup


print('All libraries imported.')
```

```python
# Assign the wikipedia page provided for the Toronto Neighborhood data

url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
toronto_neighborood_url = requests.get(url).text

#website_url    # commented out - but can be used to view full content - DJM 6/10/2020
```

As per instructions:  Use pandas, or the BeautifulSoup package, or any other way you are comfortable with to transform the data in the table on the Wikipedia page into the above pandas dataframe.

```python
# parse the doc using lxml's HTML parser

soup = BeautifulSoup(toronto_neighborood_url, 'lxml')

#print(soup.prettify())      #commented out - use to view full content - DJM 6/10/2020
```

To create the dataframe:

<p>The dataframe will consist of three columns: PostalCode, Borough, and Neighborhood</p>
<p>Only process cells that have an assigned borough. Ignore cells with a borough that is Not assigned.</p>
<p>More than one neighborhood can exist in one postal code area. For example, in the table on the Wikipedia page, you will notice that M5A is listed twice and has two neighborhoods: Harbourfront and Regent Park. These two rows will be combined into one row with the neighborhoods separated with a comma as shown in row 11 in the above table.</p>
<p>If a cell has a borough but no assigned neighborhood, then the neighborhood will be the same as the borough.</p>

```python
# locate the table of interest

table = soup.find('table', {'class':'wikitable sortable'})

#table  # commented out - use to view full content - DJM 6/10/2020
```

```python
# extract the data within the <tr> markup tags

tor_data = table.findAll('tr')

#data  # commented out - use to view full content - DJM 6/10/2020
```

```python
# Now, create a new list and store records identified as candidates within the loop:  -- DJM 6/10/2020
neighbor_table = []

# loop through each record in the data and append it to the empty list
for row in tor_data:
    neighbor_table.append([t.text.strip() for t in row.findAll('td')])

# create the new dataframe where Borough is assigned:  Ignore cells where borough is Not Assigned
df = pd.DataFrame(neighbor_table, columns = ['PostalCode', 'Borough', 'Neighborhood'])
df = df[df['Borough'] != 'Not assigned']  

# clean up the layout of the dataframe
df.drop(df.head(1).index, inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

#### create a function which retrieves the nearest venues to each Neighborhood within a 500 meter radius

```python
def getNearby_Venues(names, latitudes, longitudes, radius = 500):
    venues_list = []
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
                CLIENT_ID, CLIENT_SECRET, VERSION, lat, lng, radius, limit)
        results = requests.get(url).json()['response']['groups'][0]['items']
        venues_list.append([(
            name,
            lat,
            lng,
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'], 
            v['venue']['categories'][0]['name']) for v in results])
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    return(nearby_venues)
```

```python
# run the above function for each neighborhood  -- DJM 6/16/2020
toronto_venues = getNearby_Venues(names = toronto_borough_data['Neighborhood'],
                                 latitudes = toronto_borough_data['Latitude'],
                                 longitudes = toronto_borough_data['Longitude'])
```

### Here we will perform some simple analysis for the neighborhoods within Toronto

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add the neighborhood column back to dataframe
toronto_onehot.drop(['Neighborhood'], axis=1, inplace=True)
toronto_onehot.insert(loc=0, column='Neighborhood', value=toronto_venues['Neighborhood'])

toronto_onehot.head()
```

```python
# group rows by neighborhood & then determine the mean of the frequency of each occurence:  illustrated in prior class lab - djm 6/16/2020
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped.head()
```

```python
# Create function to return the most common venues -- djm 6/16/2020
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
tor_neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
tor_neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    tor_neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

tor_neighborhoods_venues_sorted.head()
```
